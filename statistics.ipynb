**1. What is a random variable in probability theory?**

A random variable in probability theory is a numerical description of the outcome of a random experiment. It is a function that assigns a real number to each possible outcome in the sample space of a random phenomenon. Random variables are typically denoted by capital letters, such as (X) or (Y).


**2. What are the types of random variables?**

• There are two main types of random variables:

• Discrete random variables — These take on a finite or countable number of values. An example is the number of heads when flipping a coin multiple times.

• Continuous random variables — These take on an infinite number of values within a given range. An exampleis the height of individuals in a population, which can vary continuously.

**4. What is the binomial distribution, and how is it used in probability?**

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure. It is applicable when the following conditions are met: a fixed number of trials ((n)) are conducted, each trial is independent, each trial has only two possible outcomes (success or failure), and the probability of success ((p)) remains constant for every trial.

Usage in Probability
The binomial distribution is used in probability to calculate the likelihood of obtaining a specific number of successes in a series of independent Bernoulli trials.

**5. What is the standard normal distribution, and why is it important?**

The standard normal distribution is a specific normal distribution with a mean of 0 and a standard deviation of 1, characterized by a symmetrical bell-shaped curve. It is important because it provides a universal scale (the z-score) to normalize any normal distribution, enabling the comparison of data from different distributions and allowing for the calculation of probabilities and percentiles using a common reference tool like a Z-table

Standardization:
It serves as a universal scale to analyze and compare data from any normal distribution by converting those values into z-scores.

Probability calculation:
The standard normal distribution allows statisticians to find the probability of certain values occurring, using a standard normal table (or Z-table).

Data comparison:
It makes it easy to compare data points from different normal distributions that have different means and standard deviations.

Inference and testing:
It's fundamental for many statistical procedures, including hypothesis testing and determining confidence intervals for populations.

Empirical rule:
It helps in understanding the data's spread; for example, the empirical rule states that approximately 68% of data falls within one standard deviation of the mean, 95% within two, and 99.7% within three.

**6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

The Central Limit Theorem (CLT) states that when you take sufficiently large random samples from a population, regardless of the population's original distribution, the distribution of the sample means will approximate a normal distribution. This is critical because it allows statisticians to use statistical methods that assume normality (like t-tests and confidence intervals) to make reliable inferences about a population, even when the underlying data is not normally distributed, thereby underpinning much of modern statistics.

Justifies normal-based methods:
Many powerful statistical tests and estimation techniques, such as confidence intervals and hypothesis tests, assume that the data being
analyzed is normally distributed.

Enables inference with non-normal data:
The CLT provides a mathematical justification for using these normal-based methods even when the original population data is not normally distributed.

Underpins much of statistical practice:
Because so many real-world datasets do not conform to a normal distribution, the CLT is fundamental to making reliable inferences, conducting meaningful analysis, and making decisions in fields like medicine, finance, and engineering.

**7: What is the significance of confidence intervals in statistical analysis?**

Confidence intervals provide a range of plausible values for a population parameter, indicating the precision and reliability of a sample estimate. They quantify the uncertainty inherent in sample-based data, offering more information than a single point estimate or a p-value by showing how much the true population value might vary across repeated studies. A narrower confidence interval suggests a more stable and precise estimate, while a wider one indicates greater instability and uncertainty. **8: What is the concept of expected value in a probability distribution?**

The concept of expected value is the "average" outcome of a random variable in a probability distribution, representing the long-term average if the experiment were repeated many times. It is calculated by summing the product of each possible outcome's value and its corresponding probability, and it serves as a crucial tool for decision-making, risk analysis, and understanding central tendency

**9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 50
std_dev = 5
num_samples = 1000

# Generate random numbers from a normal distribution
random_numbers = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# Compute the mean and standard deviation using NumPy
computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

print(f"Generated {num_samples} random numbers from a normal distribution.")
print(f"Theoretical Mean: {mean}, Theoretical Standard Deviation: {std_dev}")
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

# Draw a histogram to visualize the distribution
plt.hist(random_numbers, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

** 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.**
**daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**

**● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval**

**● Write the Python code to compute the mean sales and its confidence interval**



To apply the Central Limit Theorem (CLT), we would take multiple random samples from the provided daily sales data, calculate the mean for each sample, and then use the distribution of these sample means to estimate the true population mean. This method assumes the sample size is large enough (typically 30 or more) for the CLT to hold, even if the original data isn't normally distributed, allowing us to construct a 95% confidence interval for the overall average sales

```python
import numpy as np
from scipy import stats

daily_sales = [
    220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
    235, 260, 245, 250, 225, 270, 265, 255, 250, 260
]

# Convert to a NumPy array for easier calculations
sales_array = np.array(daily_sales)

# Calculate the sample mean
mean_sales = np.mean(sales_array)

# Calculate the sample standard deviation
std_dev_sales = np.std(sales_array, ddof=1)


n = len(sales_array)

standard_error = std_dev_sales / np.sqrt(n)


z_score = 1.96 # For a 95% CI, a common approximation


# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
confidence_interval_lower = mean_sales - margin_of_error
confidence_interval_upper = mean_sales + margin_of_error

print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")
```
